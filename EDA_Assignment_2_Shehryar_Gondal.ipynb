__Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.__

```python
df_Google['Rating'].value_counts()
```

```python
df_Google.info()
```

```python
df_Google.dropna(subset=['Size', 'Rating'], inplace=True)

# remove non-numeric characters from the 'Size' column
df_Google['Size'] = df_Google['Size'].str.replace('M', '').str.replace(',', '')

# convert the 'Size' column to numeric values in bytes
df_Google['Size'] = pd.to_numeric(df_Google['Size'], errors='coerce') * 1024 * 1024

# create a scatter plot of the Size and Rating columns
plt.scatter(df_Google['Size'], df_Google['Rating'], alpha=0.3)

plt.title('Size vs Rating of Apps')
plt.xlabel('Size (bytes)')
plt.ylabel('Rating')

plt.show()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_Google[cols]=df_Google[cols].str.replace(item,'')
```

__Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.__

```python
# calculate the average price by app type
df_Google['Price'] = pd.to_numeric(df_Google['Price'], errors='coerce')
avg_price_by_type = df_Google.groupby('Type')['Price'].mean()

# create a bar chart of the average prices by app type
plt.bar(avg_price_by_type.index, avg_price_by_type.values)

plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')

plt.show()
```

```python
Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.
```

```python
# Group by app and sum the installs
app_installs = df_Google.groupby('App')['Installs'].sum()

# Sort the apps by the number of installs
sorted_apps = app_installs.sort_values(ascending=False)

# Display the top 10 apps
top_10_apps = sorted_apps.head(10)
print(top_10_apps)
```
